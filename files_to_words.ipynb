```python
# 파일 불러오기

import os

path = './data/'
file_list = os.listdir(path)
file_list
```

```python
# pdf to text, 무역영어 파트 잘라서 전체 파일 합치기
import pdfplumber

english_text = ""

for file in file_list:
    pdf = pdfplumber.open(path+file)
    pages = pdf.pages
    text = ""
    for page in pages:
        sub = page.extract_text()
        text += sub

    english_text += text.split("무 역 영 어")[1]
```

```python
english_text = english_text.split('\n')
english_text
```

```python
# 정규식으로 영어만 남기기, 공백 제거
import re
 
def clean_text(data):
  text = re.sub('[^a-zA-Z]', ' ', data)
  return text
```

```python
sentence_list = list()

for i in english_text:
    text = clean_text(i).strip()
    text = ' '.join(text.split())
    sentence_list.append(text)

sentence_list = [n for n in sentence_list if n]
sentence_list
```

```python
# 띄어쓰기를 기준으로 문장 분할
word_list = []

for sentence in sentence_list:
    for i in sentence.split():
        word_list.append(i)

word_list
```

```python
# 대문자로만 구성된 단어 제거(ABC, USD 등)
word_list = [word.lower() for word in word_list if not word.isupper()]
print(len(word_list))
```

```python
# 중복 제거
word_list = list(set(word_list))
print(len(word_list))
```

```python
# 4 글자 이하인 단어 제거
word_list = [word for word in word_list if not len(word)<=4]
print(len(word_list))
```

```python
# 표제어 추출
from nltk.stem import WordNetLemmatizer

# nltk.download('wordnet')
# nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()

word_list = [lemmatizer.lemmatize(word) for word in word_list]
word_list = list(set(word_list))
len(word_list)
```

```python
# 불용어 제거
import nltk
from nltk.corpus import stopwords

# nltk.download('stopwords') # 필요시 다운로드

stop_words_list = stopwords.words('english')
word_list = [word for word in word_list if not word in stop_words_list]
len(word_list)
```

```python
# 정렬
word_list = sorted(word_list)
word_list
```

```python
# 텍스트 파일로 저장
f = open('word_list.txt', 'w')
[f.write(word+'\n') for word in word_list]
f.close()
```
